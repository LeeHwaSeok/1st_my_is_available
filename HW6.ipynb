```python
class page1:
    
    #문제의 첫 조건인 원소 중복 제거
    def __init__(self,li=[]):    
        li2=[]                  
        for a in li:            
            if a not in li2:    # 중복 원소 제거를 위해 임의의 리스트에 없는 문자만 추가한다
                li2.append(a)   
        self.li = li2           # 중복을 제거한 값을 인스턴스 변수
        
    #더하기 
    def add(self, num):       
        if num not in self.li:  #받아온 원소를 인스턴스 변수와 비교
            self.li.append(num) #추가
        
    #삭제
    def discard(self, num):      
        if num in self.li:      #받아온 원소를 인스턴스 변수 내에 있을 경우
            self.li.remove(num) #삭제
    #초기화
    def clear(self):
        self.li.clear()         #인스턴스 변수 초기화
        return 
        
    #인스턴스 변수 길이 확인
    def __len__(self):          
        cnt = 0
        for x in self.li:      #인스턴스 변수만큼 반복
            cnt = cnt +1       #카운팅
        return cnt
    
    #출력
    def __str__(self):         #__str__ >> print()와 동일
        pa = " , ".join(map(str,self.li)) # 인스턴스 변수를 문자로 변환
        return '{'+ pa +'}'    #과제 요구사항의 형식으로 출력

    #존재 여부 확인
    def __contains__(self, elem):
        if elem in self.li:         #인스턴스 변수를 파라메타와 비교하여 있을경우 참 else 거짓 출력
            return True
        else:
            return False
        
    #부분집합 A가 클 경우 
    def __ge__(self, elem):
        for x in elem.li:        #파라메타를 반복
            if not x in self.li: #인스턴스 변수안에 있을경우 
                return False     
        return True               
    
    #부분집합 B가 클 경우 
    def __le__(self, elem):
        for x in self.li:        #인스턴스 변수만큼 반복
            if not x in elem.li: #파라메타에 인자가 없을경우 
                return False
        return True    
    
    # A와 B를 비교해서 같은요소만 뽑기
    def __and__(self,elem):   
        li3 = []                 #새로운 리스트를 만든다
        for x in elem.li :       #파라메타를 반복
            if x in self.li:     #인스턴스 변수와 같으면
                li3.append(x)    #추가
        return page1(li3)   ##### 중요 클래스 안에서 객체 선언 가능
    
    # A와 B를 비교해서 같은 요소를 A에 대입
    def __iand__(self,elem):   
        li3 = []  
        for x in elem.li :
            if x in self.li:
                li3.append(x)   #위와 동일
        self.li = li3           #li3를 인스턴스 변수에 대입
        return self             #self는 class의 주소를 받아오는 개념이다 즉, self를 반환하는 것은 class를 반환하는 것과 동일
        
    #더하기
    def __or__(self,elem):     
        li3 = []
        for x in elem.li :    #파라메타를 돌고
            if x not in li3:  #없는 것만 li3에 저장
                li3.append(x)
        for y in self.li :    #객체를 돌고
            if y not in li3:  #없는 것만 저장
                li3.append(y)
        return page1(li3)     # li3를 객체화하여 반환
    
    #더하여 객체에 대입
    def __ior__(self,elem):  
        for x in elem.li :    #파라메타만 비교하여 객체에 저장
            if x not in self.li:
                self.li.append(x)
        return self # class 반환
    
    #A와 B를 비교하여 서로 없는것만 추가
    def __sub__(self,elem):
        li3 = []
        for x in elem.li :     
            if x not in self.li:  #파라메타와 객체를 비교하여 동일한 값이 없을 경우만
                li3.append(x)     #li3에 추가 
        return page1(li3)        #li3를 객체화
    
    def __isub__(self,elem):
        li3 = []
        for x in self.li :
            if x not in elem.li:
                li3.append(x)
        self.li = li3             #위와 동일하나 li3를 객체에 대입.
        return self     
                
```
